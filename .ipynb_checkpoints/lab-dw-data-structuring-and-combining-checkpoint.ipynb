```python
#### CHALLENGE 1

import pandas as pd

# Load the dataset
url1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

company_df1 = pd.read_csv(url1)
company_df2 = pd.read_csv(url2)
company_df3 = pd.read_csv(url3)
```

```python
company_df1.head()
```

```python
company_df2.head()
```

```python
company_df3.head()
```

```python
#Estandarizo nombres de atributos (columnas): hago minúsculas y reeplazo " " por "_"

company_df1.columns = [column.lower().replace(" ", "_") for column in company_df1.columns]
company_df1
```

```python
#Estandarizo nombres de atributos (columnas): hago minúsculas y reeplazo " " por "_"

company_df2.columns = [column.lower().replace(" ", "_") for column in company_df2.columns]
company_df2
```

```python
#Estandarizo nombres de atributos (columnas): hago minúsculas y reeplazo " " por "_"

company_df3.columns = [column.lower().replace(" ", "_") for column in company_df3.columns]
company_df3
```

```python
#Estandarizo nombres de columnas para que todos los dataframes tengan nombres de atributs (columnas) iguales
#Cambio nombre columna "st" por "state" en df1
company_df1.rename(columns={"st": "state"}, inplace = True)
company_df1
```

```python
#Estandarizo nombres de columnas para que todos los dataframes tengan nombres de atributs (columnas) iguales
#Cambio nombre columna "st" por "state" en df2
company_df2.rename(columns={"st": "state"}, inplace = True)
company_df2
```

```python
#Miro las columnas del primer data frame para ver como tengo que organizar las columnas en los otros dos df
#Copio el orden de las columnas para definir el orden en los otros df
company_df1.columns
```

```python
#Ordeno las columnas de df2 para que sean como las de df1
company_df2 = company_df2[[
    'customer', 'state', 'gender', 'education', 'customer_lifetime_value',
       'income', 'monthly_premium_auto', 'number_of_open_complaints',
       'policy_type', 'vehicle_class', 'total_claim_amount']]
company_df2.head()
```

```python
#Ordeno las columnas de df3 para que sean como las de df1
company_df3 = company_df3[[
    'customer', 'state', 'gender', 'education', 'customer_lifetime_value',
       'income', 'monthly_premium_auto', 'number_of_open_complaints',
       'policy_type', 'vehicle_class', 'total_claim_amount']]
company_df3.head()
```

```python
#Combino las tablas con un CONCAT (porque quiero agruparlas verticalmente, ya que 
#todas las columnas son iguales

company_concat = pd.concat([company_df1,company_df2,company_df3], ignore_index=True)
company_concat
```

```python
#Compruebo que en la nueva tabla estén todas las filas sumando las
#longitudes de las tablas individuales

len(company_df1)+len(company_df2)+len(company_df3) #12074 filas, está bien
```

```python
#Procedo a quitar el % de la columna "customer_lifetime_value" 
#Convierto los valores a numéricos con astype también

company_concat2 = company_concat.copy()

company_concat2["customer_lifetime_value"] = company_concat["customer_lifetime_value"].apply(lambda x: x.replace("%","") if isinstance(x, str) and x.endswith("%") else x)
company_concat2
```

```python
#Chequeo que haya la misma cantidad de valores nulos antes y después de hacer la modificación
```

```python
company_concat.isnull().sum()
```

```python
company_concat2.isnull().sum()
```

```python
#Limpiar la columna "number_of_open_complaints".
#Dejar solo el caracter central de cada valor (ej: en 1/2/00 dejar solo el 2)
#Utilizo str.split() para dividir la cadena por alguna condición (en nuestro caso sería el /)

company_concat3 = company_concat2.copy()
company_concat3["number_of_open_complaints"] = company_concat2["number_of_open_complaints"].apply(lambda x: x.split("/")[1] if isinstance(x, str) else x)
company_concat3
```

```python
#Estandarizo los valores posibles en la columna "gender"

company_concat3["gender"] = company_concat3["gender"].replace({"Male":"M", "Femal":"F", "female":"F", "Female":"F"})
company_concat3
```

```python
#Chequeo cuál es el porcentaje de valores nulos

porcentaje_nulos = (company_concat3.isnull().sum() / len(company_concat3))*100
porcentaje_nulos

#Hay aproximadamente 25% de valores nulos en cada columna, lo cual es mucho, por lo tanto para no eliminar todas esas
#filas, eliminamos las filas donde TODOS los valores son nulos
```

```python
#Elimino filas donde TODOS los valores SON nulos
company_concat4 = company_concat3.copy()

company_concat4 = company_concat3.dropna(how="all")
company_concat4
```

```python
#Chequeo cuantos null hay ahora en cada columna

company_concat4.isnull().sum() #Output: Vemos que en gender tenemos 122 
                               #valores nulos y en customer_lifetime_value hay 7 nulos       
```

```python
#Chequeo cuantos null hay en la columna "gender"

company_concat4["gender"].isnull().sum() 
```

```python
#Reemplazo los nulos teniendo en cuenta proporciones de M y F
import numpy as np

# Calcular las proporciones de los valores no nulos
proporciones = company_concat4['gender'].value_counts(normalize=True)
print(proporciones)
#Output: Vemos que son parecidas las cantidades, entonces aplicamos un fill basado en frecuencias.
```

```python
company_concat5 = company_concat4.copy()

# Reemplazar los valores nulos con valores aleatorios basados en las proporciones
company_concat5['gender'] = company_concat5['gender'].apply(lambda x: np.random.choice(proporciones.index, p=proporciones) if pd.isnull(x) else x)

company_concat5.isnull().sum()
```

```python
#Elimino los 7 valores nulos que quedan en la columna "customer_lifetime_value"

company_concat5 = company_concat5.dropna(subset=["customer_lifetime_value"])
company_concat5.isnull().sum()
```
